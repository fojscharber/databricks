Within this report contains an anaylsis of Cohorts 5 from 2022 biomass sampling measurements.



# Histograms

Starting on 2022-03-20 and still in production, below is a 30, 60, 90 day analysis of the Cohort's sampling histograms.

In the first 30 days, we see two distintive populations with the 480 data points observed.

Post 30 days, we still continue to see two distintive populations with 1,499 data points observed.

Post 60 days, we see a much denser population than previously that is both left and right leaning with 498 data points observed.

Post 90 days, we see a denser population yet that is more normal.


## First 30 Days Post Stocking: 2022-03-20 to 2022-04-22

```
%sql

SELECT
  sampleweight as sampled_weight, batch, to_date(date, 'MM-dd-yy') as date
FROM
  hive_metastore.aquamanager_growout_dbo.view_sampling_details
WHERE 
  batch = 'SR-C0522MxEc' AND to_date(date, 'MM-dd-yy') BETWEEN to_date('2022-03-20', 'yyyy-MM-dd') AND to_date('2022-04-22', 'yyyy-MM-dd')
ORDER BY
  date
ASC
```
